```python
""" --- User input ----------------------------- """

# 100 micron DEXMES
masterDIR = '/Users/strom-adm/Documents/Floc-Processing/2022-05-04-DEXMES-100muSand' # path to the data
saveto = 'Data-FlocProcessed/FlocCam-DEXMAS-100muSand-All.csv'

# 200 micron DEXMsS
# masterDIR = '/Users/strom-adm/Documents/Floc-Processing/2022-05-06-DEXMES-200muSand'
# saveto = 'Data-FlocProcessed/FlocCam-DEXMAS-200muSand-All.csv'

PSDfiles = 'PSD_num1_bintype_LISST.csv'    # camera file



"""  ------------------------------------------- """

# Find all of the data files and store the names 

files = sorted(glob.glob(masterDIR+"/*/*/PSD_num1_bintype_LISST.csv")) 
files
```

## Create a single dataframe and save it

```python
# start the dataframe off
dfmaster = pd.read_csv(files[0], usecols=[0,1], header=0)

for i in range(1,len(files)):
    df = pd.read_csv(files[i], usecols=[1], header=0)
    dfmaster = pd.concat([dfmaster,df.vol_muL], axis = 1)

dfmaster.to_csv(saveto, index=False)

dfmaster
```
